```python
#for trust the model
from TTS.tts.models.xtts import XttsArgs
```

```python
import torch
from TTS.api import TTS
from TTS.tts.configs.xtts_config import XttsConfig, XttsAudioConfig
from TTS.config.shared_configs import BaseDatasetConfig
from TTS.tts.models.xtts import XttsArgs  # Eksik olan bu!

# Tüm gerekli sınıfları güvenli hale getir
torch.serialization.add_safe_globals([
    XttsConfig,
    XttsAudioConfig,
    BaseDatasetConfig,
    XttsArgs
])
```

```python
# Get device
device = "cuda" if torch.cuda.is_available() else "cpu"

# Init TTS
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)

# TTS the my_voice.wav
tts.tts_to_file(
    text="Benim adım Talat.",
    speaker_wav="my_voice.wav",
    language="tr",
    file_path="output.wav"
)
```

```python
from IPython.display import Audio
# orjinal voice display
Audio("my_voice.wav")
```

```python
from transformers import pipeline
from TTS.api import TTS

# Translate
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-tr-en")
translated = translator("Benim adım Talat.")[0]['translation_text']
print("Translated text:", translated)

# 2.Speak the translated text in that language using TTS
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to("cuda" if torch.cuda.is_available() else "cpu")
tts.tts_to_file(
    text=translated,
    speaker_wav="my_voice.wav",
    language="en",
    file_path="english_v.wav"
)
```
